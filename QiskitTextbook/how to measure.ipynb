```python
from qiskit import QuantumCircuit

ckt = QuantumCircuit(1, 2)
ckt.h(0)
ckt.measure(0, 1)
ckt.x(0)
ckt.measure(0, 1)
ckt.draw()
```

```python
from qiskit.providers.aer import QasmSimulator
sim = QasmSimulator()
job = sim.run(ckt)
result = job.result()
counts = result.get_counts(ckt)

from qiskit.visualization import plot_histogram
plot_histogram(counts)
```
