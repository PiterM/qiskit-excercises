```python
from ipynb.fs.full.quantum_helpers import q_plot_histogram
from qiskit import QuantumCircuit

ckt = QuantumCircuit(12)
a = 0
b = 1
c = 2
d = 3
e = 4
f = 5
aa = 6
bb = 7
cc = 8
dd = 9
ee = 10
ff = 11
ckt.h([a, b, c, d, e, f])
ckt.barrier()
ckt.draw()
```
