```python
from ipynb.fs.full.quantum_helpers import q_plot_histogram
from qiskit import QuantumCircuit

people_number = 4
ckt = QuantumCircuit(people_number * 2)

ckt.h(range(people_number))
ckt.barrier()
ckt.draw()
```

```python

for i in range(people_number - 1):
  for j in range(people_number - i - 1):
    ckt.cx(j + i + 1, people_number + i)
  ckt.barrier()

  for k in range(people_number - i - 1):
    ckt.cx(people_number + i, people_number + i + k + 1)  
  ckt.barrier()
  
ckt.draw()
```

```python
ckt.measure_all()
q_plot_histogram(ckt)
```
