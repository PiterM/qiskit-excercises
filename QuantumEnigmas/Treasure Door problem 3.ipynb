## Quantum Enigma 001: Treasure Door
from https://learn.qiskit.org/problem-sets/quantum_enigma_problem_set/overview

```python
from ipynb.fs.full.quantum_helpers import q_plot_histogram
from qiskit import QuantumCircuit

qc = QuantumCircuit(3)

qc.h(0)
qc.h(2)
qc.cx(0, 1)
qc.barrier(0, 1, 2)
qc.cx(2, 1)
qc.x(2)
qc.cx(2, 0)
qc.x(2)
qc.barrier(0, 1, 2)
qc.swap(0, 1)
qc.x(1)
qc.cx(2, 1)
qc.x(0)
qc.x(2)
qc.cx(2, 0)
qc.x(2)

measured_qc = qc.measure_all(inplace=False)
qc.draw()
```

```python
qc.measure_all()
q_plot_histogram(qc)
```
